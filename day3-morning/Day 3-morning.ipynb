# Find most abundant gene
    1. Open t_data cile
    2. Look line by line
        3. Read FPKM and gene_name from that line
        4. if new FPKM is larger than previous FPKM then store the larger FPKM and gene_name
    5. At end print gene with the highest one

```python
file_ctab = open('/Users/cmdb/data/results/stringtie/SRR072893/t_data.ctab', 'r')
max_FPKM = 0.0
max_gene = ''
for i, line in enumerate(file_ctab):
    if i == 0:
        continue
        #print(type(line))
        #print(i, line)
    column_info = line.split()
    gene_name = column_info[9]
    FPKM = float(column_info[11])
        #print(gene_name, FPKM)
    if FPKM > max_FPKM:
        max_FPKM = FPKM
        max_gene = gene_name
print(max_gene, max_FPKM)
```

```python
file_ctab = open('/Users/cmdb/data/results/stringtie/SRR072893/t_data.ctab', 'r')

FPKMs={}
for i, line in enumerate (file_ctab):
    if i == 0:
        continue
        #print(type(line))
        #print(i, line)
    column_info = line.split()
    t_name = column_info[5]
    FPKM = float(column_info[11])
    FPKMs[t_name] = FPKM
len(FPKMs.keys() )
```

```python
for gene in FPKMs:
    if FPKMs[gene] >100:
        print(gene, FPKMs[gene])
    
    
```
