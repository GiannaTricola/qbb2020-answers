# Import and read file into DF

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../../Downloads/matrix_1kg.txt', sep = '\t')
df
```

# Include only columns with Alt Alleles

```python
df2 = df.iloc[:, 4:]
df2
```

# Calculated allelic frequency

```python
allelic_freq = []
counts = df2.sum(axis =1)
allelic_freq = counts/ (2548 * 2)


```

# Made weird looking histogram

```python
fig, ax = plt.subplots()
ax.hist(x = allelic_freq)
ax.set_ylabel('SNP')
ax.set_xlabel('Allelic Frequency')
ax.set_title('Allelic Frequency Spectrum')
```

```python
df2['Allelic Frequency'] = allelic_freq
df2
```

# Made df common variation

```python
snp1 = df2.loc[:, 'Allelic Frequency'] > 0.05
snp2 = df2.loc[:, 'Allelic Frequency'] < 0.95
snp1
snp2
snp3= snp1 & snp2
common_variation = df2.loc[snp3, :]
common_variation

```

```python
common_variation2 = common_variation.loc[:,'Allelic Frequency']
common_variation2
```

# Better looking histogram using common variation

```python
fig, ax = plt.subplots()
ax.hist(x = common_variation2)
ax.set_title('Allelic Frequency Spectrum with Common Variation')
ax.set_xlabel('Allelic Frequency')
ax.set_ylabel('SNP')
```

# PCA 

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
pca_input = df2.T
pca = PCA(n_components = 2)
pca_output = pca.fit_transform(pca_input)
pca_output
```

```python
pca_output.shape
```

```python
pca_output_df = pd.DataFrame(data = pca_output,
                            columns = ['PC1', 'PC2'])
pca_output_df
```

```python
fig, ax = plt.subplots()
ax.scatter(x = pca_output_df['PC1'], y = pca_output_df['PC2'])
ax.set_title('PCA of SNP Genotypes')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
```
