## Exercise 1 & 2: Import and read file into DF

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../../Downloads/matrix_1kg.txt', sep = '\t')
df
```

## Include only columns with allele information

```python
df2 = df.iloc[:, 4:]
df2
```

## Calculate allelic frequency

```python
allelic_freq = []
counts = df2.sum(axis =1)
samp= len(df2.columns)
allelic_freq = counts/ (samp * 2)
allelic_freq


```

## Plot Allelic Frequency Sprectrum

```python
fig, ax = plt.subplots()
ax.hist(x = allelic_freq)
ax.set_ylabel('Counts')
ax.set_xlabel('Allelic Frequency')
ax.set_title('Allelic Frequency Spectrum')
plt.show()
```

```python
df2['Allelic Frequency'] = allelic_freq
df2
```

## Exercise # 3: DataFrame common variation

```python
snp1 = df2.loc[:, 'Allelic Frequency'] > 0.05
snp2 = df2.loc[:, 'Allelic Frequency'] < 0.95
snp1
snp2
snp3= snp1 & snp2
common_variation = df2.loc[snp3, :]
common_variation

```

```python
common_variation2 = common_variation.loc[:,'Allelic Frequency']
common_variation2
```

## Plot Allelic Frequency Spectrum using Common Variation

```python
fig, ax = plt.subplots()
ax.hist(x = common_variation2)
ax.set_title('Allelic Frequency Spectrum using Common Variation')
ax.set_xlabel('Allelic Frequency')
ax.set_ylabel('Counts')
plt.show()
```

# Exercise #4 PCA 

```python
#import packages necessary for PCA
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
#Transpose data
pca_input = df2.T
#standardize data
pca_input_standardized = StandardScaler().fit_transform(pca_input)
pca_input_standardized.mean(axis=0)
pca_input_standardized.var(axis = 0)
```

```python
#run PCA with 2 components
pca = PCA(n_components = 10)
pca_output = pca.fit_transform(pca_input)
pca_output
```

```python
pca_output.shape
```

```python
pca_output_df = pd.DataFrame(data = pca_output,
                            columns = ['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8', 'PC9', 'PC10'])
pca_output_df
```

```python
fig, ax = plt.subplots()
ax.scatter(x = pca_output_df['PC1'], y = pca_output_df['PC2'])
ax.set_title('PCA of SNP Genotypes')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
plt.show()
```
