```python
import pandas as pd
import seaborn
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/cmdb/Downloads/abundance_table.tab', sep = '\t')
data
```

```python
values = data.drop('Genomic bins', axis =1)
```

```python
values.astype(float)
```

```python
plt.subplots(figsize= (10, 5))
seaborn.heatmap(values, cmap='plasma')
y_ticks = ['Staphylococcus haemolyticus', 'Enterococcus faecalis', 'Leuconostoc citreum', 'Cutibacterium avidum', 'Staphylococcus epidermidis', 'Staphylococcus aureus', 'Anaerococcus prevotii', 'Staphylococcus lugdunensis']
tick = [0,1,2,3,4,5,6,7]
plt.yticks(ticks=tick , labels=y_ticks, rotation = 'horizontal')
plt.ylabel('Species')
plt.xlabel('Samples')
plt.show()
```
